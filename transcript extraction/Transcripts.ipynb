```python
from urllib.request import urlopen
from tqdm import tqdm
from youtube_transcript_api import YouTubeTranscriptApi
import json
import re
```

## Get all videos of a specific channel
(Will not be necessary if we extract the transcripts real-time)

```python
def get_all_video_in_channel(channel_id):
    api_key = 'AIzaSyDB5tRxxmbMRGMvMfAJW2VC2mN9jAP_Tto' # Personal Api Key

    base_video_url = 'https://www.youtube.com/watch?v='
    base_search_url = 'https://www.googleapis.com/youtube/v3/search?'

    first_url = base_search_url+'key={}&channelId={}&part=snippet,id&order=date&maxResults=25'.format(api_key, channel_id)
    print(first_url)

    video_links = []
    url = first_url
    while True:
        inp = urlopen(url,timeout=1)
        resp = json.load(inp)

        for i in resp['items']:
            if i['id']['kind'] == "youtube#video":
                video_links.append(base_video_url + i['id']['videoId'])

        try:
            next_page_token = resp['nextPageToken']
            url = first_url + '&pageToken={}'.format(next_page_token)
        except:
            break
    return video_links
```

## Get channel_id by name
(Will not be necessary if we extract the transcripts real-time)

```python
# May contains errors, couldn't test due to the youtube quota.
api_key = "AIzaSyDB5tRxxmbMRGMvMfAJW2VC2mN9jAP_Tto"
channel_name = "khanacademy"
channel_extract_url = urlopen("https://www.googleapis.com/youtube/v3/channels?key="+api_key+"&forUsername="+channel_name+"&part=id",timeout=1)
channel_info = json.load(channel_extract_url)
channel_id = channel_info.items.id

# khan_id = "UC4a-Gbdw7vOaccHmFo40b9g"
vids = get_all_video_in_channel(channel_id) # Returns the urls of all videos
```

## Get all transcripts of the found videos
(Will not be necessary if we extract the transcripts real-time)

```python
video_ids = []
video_transcripts = []
failed_ids = []
for url in tqdm(vids):
    video_id = re.search(r"v=(?P<video_id>[a-zA-Z\d_\-]+)", url).group(1)
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        video_transcripts.append(transcript)
        video_ids.append(video_id)
#         with open(video_id+'.txt', 'w') as file:
#             json.dump(transcript, file)
    except:
        failed_ids.append(video_id)
print("Succesfull transcripts: {}".format(len(video_transcripts)))
print("Failed transcripts: {}".format(len(failed_ids)))
```

## Creates a JSON of the transcript given the url

```python
def url_to_json(url):
    video_id = re.search(r"v=(?P<video_id>[a-zA-Z\d_\-]+)", url).group(1)
    transcript = YouTubeTranscriptApi.get_transcript(video_id)
    return json.dumps(transcript)

# example
url = "https://www.youtube.com/watch?v=UMTeKCLvv9E&ab_channel=KhanAcademy"
url_to_json(url)
```
